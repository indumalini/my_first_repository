```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv(r'Desktop\OnlineRetail.csv',encoding='unicode_escape')
print(df)
```

```python
df.head()
```

```python
df['Description'] = df['Description'].str.strip()
df.dropna(axis=0, subset=['InvoiceNo'], inplace=True)
df['InvoiceNo'] = df['InvoiceNo'].astype('str')
df = df[~df['InvoiceNo'].str.contains('C')]
```

```python
print(df)
```

```python
basket = (df[df['Country'] =="France"]
          .groupby(['InvoiceNo', 'Description'])['Quantity']
          .sum().unstack().reset_index().fillna(0)
          .set_index('InvoiceNo'))
```

```python
basket.head()
```

```python
def encode_units(x):
    if x <= 0:
        return 0
    if x >= 1:
        return 1

basket_sets = basket.applymap(encode_units)
basket_sets.drop('POSTAGE', inplace=True, axis=1)
```

```python
records=[]
for i in range(0,532621):
    records.append([str(df.values[i,j]) for j in range(0,8)])
```

```python
for sets in records:
    print(sets)
```

```python
association_rules=apriori(records,min_support=0.5,min_confidence=0.7,min_length=2)
```

```python
association_results=list(association_rules)
```
